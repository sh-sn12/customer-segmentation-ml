```python
print(df['Revenue'].value_counts(normalize=True) * 100)
```

```python
plt.figure(1,figsize=(15,6))
n=0
for x in ['Administrative','Administrative_Duration','Informational']:
    n+=1
    plt.subplot(1,3,n)
    plt.subplots_adjust(hspace=0.5,wspace=0.5)
    sns.distplot(df[x][df['Revenue']==True],bins=30)
    plt.title('Distplot of {}'.format(x))
plt.show()
```

```python
plt.figure(1,figsize=(15,6))
sns.countplot(y='Month',data=df[df['Revenue']==True])
plt.show()
```

```python
plt.figure(1,figsize=(15,6))
n=0
for cols in ['Administrative_Duration','Informational_Duration','ProductRelated_Duration']:
    n+=1
    plt.subplot(1,3,n)
    sns.set_style('whitegrid')
    plt.subplots_adjust(hspace=0.5,wspace=0.5)
    sns.violinplot(x=df['Revenue'],y=df[cols])
    plt.ylabel(cols)
    plt.title('Violin Plot ')
plt.show()
```

```python
numeric_cols = ['Administrative', 'Informational', 'ProductRelated',
                'Administrative_Duration', 'Informational_Duration', 'ProductRelated_Duration',
                'BounceRates', 'ExitRates', 'PageValues', 'SpecialDay']

df[numeric_cols].hist(figsize=(15,12), bins=30)
plt.suptitle("Distribution of Numeric Features")
plt.show()
```

```python
sns.countplot(x='Revenue', data=df)
plt.title("Revenue Distribution")
plt.show()
```

```python
sns.countplot(x='VisitorType', data=df)
plt.title("Visitor Type Distribution")
plt.show()
```

```python
sns.countplot(x='Month', data=df, order=df['Month'].value_counts().index)
plt.title("Sessions by Month")
plt.show()
```

```python
sns.countplot(x='Weekend', data=df)
plt.title("Weekend vs Weekday Sessions")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm', annot=False)
plt.title("Correlation Heatmap")
plt.show()
```

```python
for col in ['PageValues', 'ExitRates', 'ProductRelated', 'Administrative_Duration']:
    plt.figure(figsize=(6,4))
    sns.boxplot(x='Revenue', y=col, data=df)
    plt.title(f"{col} vs Revenue")
    plt.show()
```

```python
for col in ['Month', 'VisitorType', 'Weekend']:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, hue='Revenue', data=df)
    plt.title(f"{col} vs Revenue")
    plt.show()
```

```python
for col in ['ProductRelated', 'Administrative_Duration', 'PageValues']:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.title(f"Outliers in {col}")
    plt.show()
```
